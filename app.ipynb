### Part B

```python
# Place your dataframes into a dictionary for cleaner iteration
df_dict = {
    'lieux_df': lieux_df, 
    'caract_df': caract_df, 
    'usagers_df': usagers_df, 
    'vehicules_df': vehicules_df
}

for df_name, df in df_dict.items():
    print(f"\n{df_name.upper()}:")
    nb_total_missing = 0
    total_cells = df.shape[0] * df.shape[1]
    
    for col in df.columns:
        # 1. Standard missing values (NaN/None)
        is_null = df[col].isnull()
        
        # 2. String 'nan' placeholders
        is_string_nan = df[col].astype(str).str.strip().str.lower() == 'nan'
        is_string_null = df[col].astype(str).str.strip().str.lower() == 'null'
        is_invalid_string = is_string_nan | is_string_null
        
        # 3. Numeric -1 placeholders
        is_minus_one = df[col] == -1
        
        # Combine using bitwise OR (|) to ensure each cell is only counted ONCE
        col_missing_mask = is_null | is_invalid_string | is_minus_one
        col_missing = col_missing_mask.sum()
        
        missing_percentage = (col_missing / len(df)) * 100
        print(f"  {col}: {missing_percentage:.2f}%")
        
        nb_total_missing += col_missing

    # Calculate total overall missing percentage accurately
    total_missing_percentage = (nb_total_missing / total_cells) * 100
    print(f"\nPercentage of total missing values: {total_missing_percentage:.2f}%")
    print("-" * 50)
```

```python
import pandas as pd
import numpy as np
```

```python
def standardize_missing_values(df: pd.DataFrame) -> pd.DataFrame:
    cleaned = df.copy()
    cleaned = cleaned.replace({-1: np.nan, "-1": np.nan, "nan": np.nan, "null": np.nan, "": np.nan})
    for column in cleaned.columns:
        cleaned[column] = cleaned[column].astype("string").str.strip()
        cleaned[column] = cleaned[column].replace({"<NA>": np.nan})
    return cleaned


def to_int(series: pd.Series) -> pd.Series:
    return pd.to_numeric(series, errors="coerce").astype("Int32")


def to_float(series: pd.Series) -> pd.Series:
    return pd.to_numeric(series, errors="coerce")


def save_clean_table(df: pd.DataFrame, filename: str) -> None:
    output_path = DATA_DIR / filename
    df.to_csv(output_path, sep=";", index=False, na_rep="")


caract_clean = standardize_missing_values(caract_df)
caract_clean = caract_clean.dropna(subset=["col", "adr"]).copy()
for column in ["Num_Acc", "jour", "mois", "an", "lum", "agg", "int", "atm", "col"]:
    if column in caract_clean.columns:
        caract_clean[column] = to_int(caract_clean[column])
for column in ["lat", "long"]:
    if column in caract_clean.columns:
        caract_clean[column] = to_float(caract_clean[column])
caract_clean = caract_clean.drop_duplicates().copy()


lieux_clean = standardize_missing_values(lieux_df)
for column in ["Num_Acc", "catr", "v1", "circ", "nbv", "vosp", "prof", "pr", "pr1", "plan", "larrout", "surf", "infra", "situ", "vma"]:
    if column in lieux_clean.columns:
        lieux_clean[column] = to_int(lieux_clean[column])
lieux_clean = lieux_clean.drop(columns=["v2", "lartpc"], errors="ignore")
lieux_clean = lieux_clean.dropna(subset=["circ", "surf", "vma"]).copy()
lieux_clean = lieux_clean[lieux_clean["vma"].between(30, 130, inclusive="both")].copy()
lieux_clean = lieux_clean.drop_duplicates().copy()


vehicules_clean = standardize_missing_values(vehicules_df)
for column in ["Num_Acc", "senc", "catv", "obs", "obsm", "choc", "manv", "motor", "occutc"]:
    if column in vehicules_clean.columns:
        if column == "occutc":
            vehicules_clean[column] = to_float(vehicules_clean[column])
        else:
            vehicules_clean[column] = to_int(vehicules_clean[column])
vehicules_clean = vehicules_clean.drop(columns=["occutc"], errors="ignore")
vehicules_clean = vehicules_clean.dropna(subset=["obs", "obsm", "choc", "manv", "motor"]).copy()
vehicules_clean = vehicules_clean.drop_duplicates().copy()


usagers_clean = standardize_missing_values(usagers_df)
usagers_clean = usagers_clean.drop(columns=["secu2", "secu3", "etatp"], errors="ignore")
usagers_clean["locp"] = usagers_clean["locp"].fillna("0")
for column in ["Num_Acc", "place", "catu", "grav", "sexe", "an_nais", "trajet", "secu1", "locp"]:
    if column in usagers_clean.columns:
        usagers_clean[column] = to_int(usagers_clean[column])
usagers_clean = usagers_clean.dropna(subset=["an_nais"]).copy()
usagers_clean = usagers_clean.drop_duplicates().copy()


save_clean_table(caract_clean, "clean_caract-2024.csv")
save_clean_table(lieux_clean, "clean_lieux-2024.csv")
save_clean_table(vehicules_clean, "clean_vehicules-2024.csv")
save_clean_table(usagers_clean, "clean_usagers-2024.csv")

print("Cleaned CSV files saved in the data folder:")
print(f"- clean_caract-2024.csv: {caract_clean.shape[0]} rows, {caract_clean.shape[1]} columns")
print(f"- clean_lieux-2024.csv: {lieux_clean.shape[0]} rows, {lieux_clean.shape[1]} columns")
print(f"- clean_vehicules-2024.csv: {vehicules_clean.shape[0]} rows, {vehicules_clean.shape[1]} columns")
print(f"- clean_usagers-2024.csv: {usagers_clean.shape[0]} rows, {usagers_clean.shape[1]} columns")
```
